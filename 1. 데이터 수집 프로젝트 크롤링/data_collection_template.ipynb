# [데이터 명세]

## [코로나 확진자]

### 1.   데이터 정의

매일 발생하는 코로나 확진자의 데이터 입니다.

2020년 1월 20일에 첫 환자가 발생하였고, 매일 17:00을 기준으로 데이터가 업로드된다.

### 2. 수집 방법





* 웹 API: <u>requests</u>의 <u>get</u>을 이용하여 원하는 parameter를 보내여 원하는 datatime의 데이터를 가져옴.

* xml.etree.ElementTree을 이용하여 모든 xml데이터를 가져온 뒤 데이터가 들어있는 첫번째 아이템```item = items[0]```에 접근후 <u>기준날자</u>```statusDt = item.findtext("statusDt")```와 <u>총 감염자수</u>```hPntCnt = item.findtext("hPntCnt")```를 가져옴.

### 3.   수집 경로

공공데이터포털(보건복지부_코로나19 감염현황 총괄 통계): https://www.data.go.kr/data/15098780/openapi.do?recommendDataYn=Y#tab_layer_detail_function

API: http://apis.data.go.kr/1352000/ODMS_COVID_02/callCovid02Api

공공데이터포털(보건복지부_코로나19 시도 발생현황): https://www.data.go.kr/tcs/dss/selectApiDataDetailView.do?publicDataPk=15098776

API: http://apis.data.go.kr/1352000/ODMS_COVID_04/callCovid04Api

### 4.   데이터의 논리적 구조

확진자 수: ```<hPntCnt>확진자 수(정수)</hPntCnt>```

데이터의 날짜: ```<statusDt>datetime(시계열)</statusDt>```

## [의약품 사용 통계]

### 1.   데이터 정의

'보건의료빅데이터개방시스템'에서 제공하는 의약품 사용량을 월별로 집계한 통계입니다.
2010년 1월부터 2023년 3월까지의 데이터가 매월단위로 구성되있습니다.

### 2. 수집 방법





* '보건의료빅데이터개방시스템'에서 제공하는 <u>xls</u>파일을 받아옵니다.
* <u>selenium</u>의 <u>webdriver.Chrome</u>를 이용하여 원하는 시기를 넣어 <u>xls</u>파일을 받아옵니다. 

### 3.   수집 경로

'보건의료빅데이터개방시스템'의 의약품 사용 통계

약국 직접/처방 조제량,   약효분류별 데이터를 받는다.

http://opendata.hira.or.kr/op/opc/olapMsupInfo.do

### 4.   데이터의 논리적 구조

* 처방조제수는 월별기간을 main_index로 환자수, 수량, 사용금액을 sub_index로 
    
    
    직접조제, 처방조제를 column으로 구성된 연속형 데이터 구조
* 약효분류별는 월별기간을 main_index로 판매량과 판매액을 sub_index로

    '해열,진통,소염제', '진해거담제'를 column으로 구성된 연속형 데이터 구조

### 5.   데이터의 물리적 저장 구조

이미지는 엑셀 데이터의 일부분입니다.
