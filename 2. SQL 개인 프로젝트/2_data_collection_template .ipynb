# [데이터 명세]

## [spotify의 음악 데이터]

### 1.   데이터 정의

spotify의 API에서 제공하는 음악차트와 그 차트의 track_id와 artist_id를 이용하여 트렉의 정보와 아티스트 데이터를 불러옴

### 2. 수집 방법





* 웹 API: spotify에서 제공하는 <u>spotify</u>를 이용하여 원하는 parameter를 보내여 원하는 datatime의 데이터를 가져옴.

* spotipy.Spotify를 이용하여 데이터를 json을 받아옴 

### 3.   수집 경로

* spotipy에서 제공하는 웹API 패키지인 spotipy를 이용하여 데이터를 수집함
