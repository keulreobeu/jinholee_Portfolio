데이터에 다양한 실수가 존재하는데, 이 값이 무엇을  뜻하는지 확인하기.
-------------------------------------
케글에서 제공하는 paper에서 라만 분광법을 사용하여 얻은 스팩트럼이며,

단위유전자가 많을수록 값이 높아 진다.

단위유전자의 갯수를 구하는 공식은 다음과 같다.

$$ bias(A_w, T_x, G_y, C_z) = \frac{1}{4^k} \cdot \frac{k!}{{w! \cdot x! \cdot y! \cdot z!}} $$
